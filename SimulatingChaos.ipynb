```python
import numpy as np
```

```python
def logistic_map(r, x):
    return r * x * (1 - x)
```

```python
def get_stable_orbit(r, iterations=1000, last_n=100):
    x = 0.5
    for _ in range(iterations - last_n):
        x = logistic_map(r, x)

    orbit = []
    for _ in range(last_n):
        x = logistic_map(r, x)
        orbit.append(round(x, 6))

    return set(orbit)
```

```python
def find_bifurcation(target_period, low_r, high_r, tolerance=1e-7):
    while (high_r - low_r) > tolerance:
        mid_r = (low_r + high_r) / 2
        period = len(get_stable_orbit(mid_r))

        if period < target_period:
            low_r = mid_r
        else:
            high_r = mid_r

    return mid_r
```

```python
print("Finding bifurcation points")

r1 = find_bifurcation(2, 2.9, 3.1)
print(f"r1 (split to 2): {r1:.6f}")

r2 = find_bifurcation(4, 3.4, 3.5)
print(f"r2 (split to 4): {r2:.6f}")

r3 = find_bifurcation(8, 3.54, 3.55)
print(f"r3 (split to 8): {r3:.6f}")
```

```python
delta = (r2 - r1) / (r3 - r2)
print("-" * 30)
print(f"Calculated Feigenbaum Delta: {delta:.4f}")
print("Official Constant: 4.6692")
```
